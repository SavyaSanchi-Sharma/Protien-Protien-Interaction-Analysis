```python
import pandas as pd 
import numpy as np
```

```python
d = "../data/structural"
fs = ["Test_60_17D.csv", "Test_315_17D.csv", "Train_335_17D.csv"]
dfs = [pd.read_csv(f"{d}/{f}").assign(src=f) for f in fs]
df = pd.concat(dfs, ignore_index=True)
df = df.sort_values(["src", "PDB", "Chain", "ResIdx"])
seqs = df.groupby(["src", "PDB", "Chain"], as_index=False).agg(seq=("AA", "".join), n=("AA", "size"))
seqs.to_csv(f"{d}/protein_chain_sequences.csv", index=False)
seqs
```
